- ### Alterando o espaço de cores

  - Como Matplotlib não realiza conversões complexas de espaço de cor `(como converter de BGR para RGB, ou RGB para HSV)` de forma nativa, então utilizei a biblioteca OpenCV e o Matplotlib apenas para exibir os resultados.

```python
!pip install opencv-python
```

```python
import cv2
```

```python
# Lendo a imagem com OpenCV
photos_bgr = cv2.imread("/content/rio.jpeg")
```

```python
# Alterando o espaço de cor de BGR para RGB
photos_rgb = cv2.cvtColor(photos_bgr, cv2.COLOR_BGR2RGB)
```

```python
# Exibindo a imagem
plt.imshow(photos_rgb)
plt.axis("off")
plt.show()
```

```python
photos_gray = cv2.cvtColor(photos_rgb, cv2.COLOR_BGR2GRAY)

plt.imshow(photos_gray, cmap="gray")
plt.axis("off")  # Remove os eixos para melhor visualização
plt.show()
```
